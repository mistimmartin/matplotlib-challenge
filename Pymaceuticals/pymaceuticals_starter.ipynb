```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
read_clinical = pd.read_csv(clinical_trial_data_to_load)
read_mouse = pd.read_csv(mouse_drug_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(read_clinical, read_mouse, on="Mouse ID",how="left")


# Display the data table for preview
combined_data
```

## Tumor Response to Treatment

```python
# # Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# tumor_volume = combined_data.groupby(["Drug","Time"]).mean()["Tumor Volume (mm3)"]

# # Convert to DataFrame
# average_tumor_volume=pd.DataFrame(tumor_volume)
# # Preview DataFrame
```

```python
tumor_volume = combined_data.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
average_tumor_volume=pd.DataFrame(tumor_volume)
average_tumor_volume.reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

# Preview DataFrame

```

```python
tumor_volume_standarderror = combined_data.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]
sem_tumor_volume=pd.DataFrame(tumor_volume_standarderror)
sem_tumor_volume.reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
```

```python
average_tumor_volume=average_tumor_volume.reset_index()
pivot_average_tumor_volume = average_tumor_volume.pivot(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]
pivot_average_tumor_volume
```

```python
sem_tumor_volume=sem_tumor_volume.reset_index()
pivot_sem_tumor_volume = sem_tumor_volume.pivot(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]
pivot_sem_tumor_volume
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(pivot_average_tumor_volume.index,pivot_average_tumor_volume["Capomulin"],yerr=pivot_sem_tumor_volume["Capomulin"],
             color="r",marker="o",linestyle="dashed")
plt.errorbar(pivot_average_tumor_volume.index,pivot_average_tumor_volume["Infubinol"],yerr=pivot_sem_tumor_volume["Infubinol"],
             color="b",marker="s",linestyle="dashed")
plt.errorbar(pivot_average_tumor_volume.index,pivot_average_tumor_volume["Ketapril"],yerr=pivot_sem_tumor_volume["Ketapril"],
             color="g",marker="v",linestyle="dashed")
plt.errorbar(pivot_average_tumor_volume.index,pivot_average_tumor_volume["Placebo"],yerr=pivot_sem_tumor_volume["Placebo"],
             color="black",marker="x",linestyle="dashed")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(["Capomulin","Infubinol","Ketapril","Placebo"])
ax = plt.axes()        
ax.yaxis.grid()
plt.show()
# Save the Figure
plt.savefig("Charts/Tumor Response to Treatment")
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

# Convert to DataFrame

# Preview DataFrame
```

```python
survival_rates = combined_data.groupby(["Drug","Timepoint"]).count()
survival_rates_count=pd.DataFrame(survival_rates)
survival_rates_final=survival_rates_count.drop(columns=["Mouse ID", "Tumor Volume (mm3)"])
survival_rates_final=survival_rates_final.rename(columns={"Metastatic Sites":"Mouse Count"})
survival_rates_final.reset_index()
```

```python
survival_rates_sem = combined_data.groupby(["Drug","Timepoint"]).sem()
survival_rates_clean=pd.DataFrame(survival_rates_sem)
sem_survival_rates=survival_rates_clean.drop(columns=["Mouse ID", "Tumor Volume (mm3)"])
sem_survival_rates=sem_survival_rates.rename(columns={"Metastatic Sites":"Mouse Count"})
sem_survival_rates.reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame
```

```python
survival_rates_final = survival_rates_final.reset_index()
pivot_survival_rates_final = survival_rates_final.pivot(index="Timepoint",columns="Drug")["Mouse Count"]
pivot_survival_rates_final.head()
```

```python
sem_survival_rates = sem_survival_rates.reset_index()
pivot_survival_rates_sem = sem_survival_rates.pivot(index="Timepoint",columns="Drug")["Mouse Count"]
pivot_survival_rates_sem
```

```python
# Generate the Plot (Accounting for percentages)

# Save the Figure

# Show the Figure
plt.show()
```

```python
plt.errorbar(pivot_survival_rates_final.index,(pivot_survival_rates_final["Capomulin"]/25*100),color="r",marker="o",linestyle="dashed")
plt.errorbar(pivot_survival_rates_final.index,(pivot_survival_rates_final["Infubinol"]/25*100),color="b",marker="s",linestyle="dashed")
plt.errorbar(pivot_survival_rates_final.index,(pivot_survival_rates_final["Ketapril"]/25*100),color="g",marker="v",linestyle="dashed")
plt.errorbar(pivot_survival_rates_final.index,(pivot_survival_rates_final["Placebo"]/25*100),color="black",marker="x",linestyle="dashed")
plt.title("Survival Rate During Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Survival Rate")
plt.legend(["Capomulin","Infubinol","Ketapril","Placebo"])
ax = plt.axes()        
ax.yaxis.grid()
plt.savefig("Charts/Survival Rate During Treatment")
plt.show()
```
